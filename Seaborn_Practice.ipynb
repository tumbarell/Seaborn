**Seaborn** is a Python library that allows to plot graphs that help us extract useful insights from data.

Let's use the [Iris Dataset](https://www.kaggle.com/datasets/uciml/iris) for visualization purposes.

# Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

# Getting the data in pandas DataFrame Format

```python
# url where the data set is available
csv_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

#The csv file at the UCI repository does not contain the variable names. They are located in a separate file.
col_names = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Species']

iris_data = pd.read_csv(csv_url, names = col_names)
print(f'Number of lines: {iris_data.shape[0]} Number of columns: {iris_data.shape[1]}')
iris_data.head()
```

The *Sepal_Length* and *Sepal_Width* are the length and width of the sepals of an Iris flower, calculated in centimeters. Similarly, *Petal_Length* and *Petal_Width* are the length and width of the petals of the Iris flower, measured in centimeters. Species indicate the species of the flower. There are three types of Iris species: **Iris setosa**, **Iris virginica**, and **Iris versicolor**.

#Lineplot

In this plot, the relationship between two variables (x and y) is shown by plotting them on a 2D graph and connecting them.

There are several uses for line plots, including studying dynamic variables (that change over time), finding patterns, and tracking trends. We can generate a line plot using the lineplot function.

```python
#Creating a line plot between Sepal Length and Sepal Width of all the species
sns.lineplot(data=iris_data, x = 'Sepal_Length', y = 'Sepal_Width')
```

```python
#We can also plot separate line plots for each of the species.
sns.lineplot(data=iris_data, x = 'Sepal_Length', y = 'Sepal_Width', hue = 'Species')
```

#Scatterplot

Scatter plots are similar to line plots but preferred for static variables (that don’t change over time).
Basically, it plots all the points on a 2D graph.

Scatter plots are used to determine the correlation between two variables. We can use it via the scatterplot function.

```python
#Generating a Scatter Plot of all the data points.
sns.scatterplot(data=iris_data, x = 'Sepal_Length', y = 'Sepal_Width')
```

```python
#We can also distinguish the points by their species
sns.scatterplot(data=iris_data, x = 'Sepal_Length', y = 'Sepal_Width', hue = 'Species')
```

```python
#Plotting histograms for each species
sns.histplot(data=iris_data, x='Sepal_Length',
             kde=True,
             bins = 20,
             #bins = [1000, 10000],
             #binwidth=10,
             #binrange=(30,60),
             stat= 'probability',# 'count', 'density',
             #fill=False, element='step,
             #cumulative=True,
             hue='Species',
             #element = 'step'#'step', 'poly',
             #multiple =  'stack'  #'layer', 'stack', 'fill','dodge'
             #bivariate
             #x='xcolumn', y='ycolumn', data=data (heatmap)
             #cbar=True,
             #styling
             #shrink=0.8,
             #color='indigo',
             #palette='bone'
             )
```

#Probability Density Function (PDF)

PDF calculates the probability that a random variable will be found within a range of values. The purpose of this method is to determine which distribution a variable belongs to.

By revealing which distribution the variable belongs to, we can select the most effective machine-learning model to run on the dataset to obtain the most accurate results. We use kernel density estimation to calculate the probability density function. We can plot it using the kdeplot function.

```python
#Plotting the probability density function for Petal Length of all species
sns.kdeplot(data = iris_data, x = 'Petal_Length')
```

```python
#Plotting the probability density function for Petal Length for individual species
sns.kdeplot(data = iris_data, x = 'Petal_Length', hue='Species')
# Iris-Setosa’s PDF does not overlap with other species.
# Iris-Setosa can therefore be distinguished easily from other species.
```

#Boxplot

Box plot provides us with five insights about the data which are:

    Lower Extreme — Tells us the lowest value in the data set
    Upper Extreme — Provides us with the highest value in the dataset

The Lower Extreme and Upper Extreme are useful in detecting outliers.

    Upper Quartile — Gives us the 75th percentile of the dataset which is the value at which 75 percent of the data falls (when in ascending order).
    Lower Quartile — Gives us the 25th percentile of the dataset which is the value at which 25 percent of the data falls (when in ascending order).

The box (Lower Quartile to Upper quartile) is called the Interquartile range.

    Median — Provides us with the median of the dataset

```python
#Plotting boxplot for individual species
sns.boxplot(data = iris_data, y = 'Petal_Length', x = 'Species')
```

#Violinplot

Violin plots are similar to boxplots but also indicate the variable’s probability density function. This gives it the appearance of a violin. We can plot it using violinplot function.

```python
#Plotting violinplot for individual species
sns.violinplot(data = iris_data, y = 'Petal_Length',x = 'Species')
```

#Pairplot

Pair plot allows us to plot pairwise scatterplots for all non-categorical variables. For the same variables on the x-axis and y-axis, we get the histogram/probability density function of the variable.

Since it graphs all the plots together, it becomes very easy to analyze every feature and its correlation with other features.

The use of pair plots is not recommended when the dataset contains a large number of features because they require a considerable amount of time to plot. We can graph a pair plot using the pairplot function.

```python
%%time
#Plotting pairplot
sns.pairplot(data = iris_data)
```

```python
%%time
#Plotting the pair plots for individial species
sns.pairplot(data = iris_data, hue = 'Species')
```

#Heatmap

Heatmaps are generally used to study numerical variables’ correlation. Each cell is given a color that represents the correlation between the two variables.

Colors with a dark hue indicate a high positive correlation between the variables, whereas colors with a lighter hue indicate a high negative correlation between the variables. We can plot a heatmap using the heatmap function.

```python
#Plotting the heatmap
sns.heatmap(data = iris_data.corr(), annot = True, cmap = "GnBu")
```

From the heatmap, we noticed that the pairs that have a high correlation are:

    Petal Length and Sepal Length
    Sepal Length and Petal Width
    Petal Length and Petal Width

#Jointplot

Here we plot the scatter plot and the histogram/probability density function of two variables. The joint plot can be used for both univariate and bivariate analyses.


```python
#Plotting joint plot of Sepal Width and Petal Length
sns.jointplot(data = iris_data, x = 'Petal_Length', y = 'Sepal_Width')
```

```python
#Plotting joint plot for individual species
sns.jointplot(data = iris_data, x = 'Petal_Length', y = 'Sepal_Width', hue = 'Species')
```

#Rugplot

Similar to the jointplot, the rugplot can be used for both univariate and bivariate analyses. It shows the marginal distribution of variables on the x- and y-axes.

```python
#Plotting rug plot for only Petal length for individual species
sns.rugplot(data = iris_data, x = 'Petal_Length', hue = 'Species')
```

```python
#Plotting rug plot for only Petal length and Sepal Length for individual species
sns.rugplot(data = iris_data, x = 'Petal_Length', y = 'Sepal_Length', hue = 'Species')
```
